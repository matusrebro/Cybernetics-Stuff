```python
# importing modules
import numpy as np
import matplotlib.pyplot as plt
from control_systems import lin_model
```

```python
# linear system 1 - transfer function SISO system definition
num = [6, 1]
den = [1, 2, 1]
model1 = lin_model([num, den])
```

```python
# step response simulation
t  = np.arange(0, 10, 0.1)
u = np.ones_like(t)
x0 = [0, 0]
x, y = model1.simulation(x0, t, u)
plt.figure()
plt.plot(t, y)
```

```python
# frequency response
omega = np.logspace(-3, 3, 100)
mag, phase, nyquist = model1.freq_response(omega)
```

```python
# linear system 2 - state space MIMO system definition
A = [[-2, -1], [1, 0]]
B = [[1, 1], [0, 0]]
C = [[1, 2], [6, 2]] 
D = [[0, 0], [0, 0]]
model2 = lin_model([A, B, C, D])
```

```python
# model 2 step response
t  = np.arange(0, 10, 0.1)
u = np.ones([len(t), 2])
x0 = [0, 0]
x, y = model2.simulation(x0, t, u)
plt.figure()
plt.plot(t,y[:,0],label='y1')
plt.plot(t,y[:,1],label='y2')
plt.legend()
```

```python
# model 2 frequency response
omega = np.logspace(-3, 3, 100)
mag, phase, nyquist = model2.freq_response(omega)
```
